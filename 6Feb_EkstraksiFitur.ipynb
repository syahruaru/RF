## Import Library

```python
import openbcilib as ob
from importlib import reload
import json
import os
from scipy.io import savemat, loadmat
import numpy as np
```

## Load EMG

```python
# Opening JSON file
f = open('mytrialstart4NOV.json',)
ts_str = json.load(f)
ts_dict = json.loads(ts_str)

# Update dictionary trial start yang baru
trial_starts = ts_dict
```

```python
folder = '../mat_files/2/'

index_start_frame = 0
index_stop_frame = 320

for file in os.listdir(folder):
    emg_movement = loadmat(folder+file)
    emg = emg_movement['emg_movement']
    
    R = np.empty((4,10), float)
    for i in range(emg.shape[0]):
        print("Channel:", i)
        section_num =10
        RMS = []
    
        for j in range(section_num):
            rms_ = np.sqrt(np.mean(emg[i][index_start_frame+index_stop_frame*j:index_stop_frame*(j+1)]**2))
       
            RMS.append(rms_)
            print('Segmen' + str(j) + " : ", rms_)
        R[i,] = RMS
        
#         filename = '../mat_files/3/rms/' + file
#         dekstraksi_fitur = {'rms': R, 'label': 'ekstraksi fitur rms'}
#         savemat(filename, dekstraksi_fitur)
        
print('selesai') 
```

```python
folder = '../mat_files/2/'

for file in os.listdir(folder):
    emg_movement = loadmat(folder+file)
    emg = emg_movement['emg_movement']
    
    I = np.empty((4,10), float)
    for i in range(emg.shape[0]):
        print("Channel:", i)
        section_num =10
        iemg = []
    
        for j in range(section_num):
            iEMG = np.sum(np.abs(emg[i][index_start_frame+index_stop_frame*j:index_stop_frame*(j+1)]))
       
            iemg.append(iEMG)
            print('Segmen' + str(j) + " : ", iEMG)
        I[i,] = iemg       
        
#         filename = '../mat_files/3/iemg/' + file
#         dekstraksi_fitur = {'iemg': I, 'label': 'ekstraksi fitur iemg'}
#         savemat(filename, dekstraksi_fitur)
                   
print('selesai') 
```

```python
folder = '../mat_files/2/'

for file in os.listdir(folder):
    emg_movement = loadmat(folder+file)
    emg = emg_movement['emg_movement']
    
    S = np.empty((4,10), float)
    for i in range(emg.shape[0]):
        print("Channel:", i)
        section_num =10
        SSI = []
    
        for j in range(section_num):
            ssi = np.sum(emg[i][index_start_frame+index_stop_frame*j:index_stop_frame*(j+1)]**2)
          
            SSI.append(ssi)
            print('Segmen' + str(j) + " : ", ssi)
        S[i,] = SSI
        
#         filename = '../mat_files/3/ssi/' + file
#         dekstraksi_fitur = {'SSI': S, 'label': 'ekstraksi fitur SSI'}
#         savemat(filename, dekstraksi_fitur)
        
print('selesai') 
```

## MEMBUAT DATASET

```python
folder = '../mat_files/3/rms/'

AA_RMS =[]
for file in os.listdir(folder):
    fitur = loadmat(folder+file)
    emg_fitur = fitur['rms']
    AA_RMS.append(emg_fitur)
    
folder = '../mat_files/3/iemg/'

AA_IEMG =[]
for file in os.listdir(folder):
    fitur = loadmat(folder+file)
    emg_fitur = fitur['iemg']
    AA_IEMG.append(emg_fitur)
    
folder = '../mat_files/3/ssi/'

AA_SSI =[]
for file in os.listdir(folder):
    fitur = loadmat(folder+file)
    emg_fitur = fitur['SSI']
    AA_SSI.append(emg_fitur)
    
dataset = np.hstack([AA_RMS, AA_IEMG, AA_SSI])

# filename = '../mat_files/5/' + 'dataset6Output.mat'
# ddataset = {'dataset': dataset, 'label': 'dataset'}
# savemat(filename, ddataset)
```
